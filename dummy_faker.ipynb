```python
import pandas as pd
import random
from faker import Faker
fake = Faker('id_ID')
```

```python
df_city = pd.read_excel("city.xlsx")
```

```python
df_city
```

```python
loc_id_options = df_city["kota_id"]
loc_id_options
```

```python
n = 100
name_list = []
phone_list = []
address_list = []

for i in range(0, n):
    name = fake.name()
    name_list.append(name)

    phone = fake.phone_number()
    phone_list.append(phone)

    address = fake.street_address()
    address_list.append(address)
```

```python
loc_list = []
for i in range(0, n):
    x = random.randint(0, len(loc_id_options)-1)
    loc_list.append(loc_id_options[x])
```

```python
df_user = pd.DataFrame()
df_user['user_id'] = range(1, n+1)
df_user['name'] = name_list
df_user['phone'] = phone_list
df_user['address'] = address_list
df_user['location_id'] = loc_list
```

```python
df_user.head()
```

```python
df_user.to_csv("users_table.csv", index=False)
```

```python
df_car = pd.read_excel("car_product.xlsx")
df_car
```

```python
transmission_options = ["m", "at"]

transmission_list = []
for i in range(0, 50):
    x = random.randint(0, 1)
    transmission_list.append(transmission_options[x])

df_car['transmission'] = transmission_list
df_car.head()
```

```python
# create random description

colour_options = ["red", "yellow", "blue", "dark blue", "grey", "black","white"]

desc_list = []
for i in range(0, 50):
    x = random.randint(0, len(colour_options)-1)
    desc_list.append(colour_options[x])

df_car['description'] = desc_list
df_car.head()
```

```python
df_car.to_csv("cars_table.csv", index = False)
```

```python
n = len(df_car)

ad_id_list = []
title_list =[]
car_id_list =[]
user_id_list =[]
loc_id_list = []

available_cars = list(df_car['product_id'])
for i in range(1, n+1):

    u = random.randint(0, len(df_user)-1)
    user = df_user.iloc[u]

    # c = random.randint(0, len(df_car)-1)
    c = random.choice(available_cars)
    # remove the randomly picked car_id from list
    available_cars.remove(c)
    l = random.randint(0, len(df_city)-1)

    car = df_car[df_car['product_id']==c]
    loc = df_city.iloc[l]
    title1 = f"Jual Mobil {car.model.item()} Tahun {car.year.item()} di {loc.nama_kota}"
    title2 = f"Jual Cepat {car.brand.item()} Tahun {car.year.item()} di {loc.nama_kota}"
    title3 = f"Jual Murah {car.brand.item()} Tahun {car.year.item()} di {loc.nama_kota}"
    titles = [title1, title2,title3]
    t = random.choice(titles)
    print(loc.nama_kota)
    ad_id_list.append(i)
    title_list.append(t)
    car_id_list.append(car['product_id'].item())
    user_id_list.append(user['user_id'])
    loc_id_list.append(loc['kota_id'])
```

```python
date_list = []
for i in range(0, 50):
    fake_date = fake.date_between(start_date='-1y', end_date='today')
    date_list.append(fake_date)
```

```python
df_ads = pd.DataFrame()
df_ads['ad_id'] = ad_id_list
df_ads['title'] = title_list
df_ads['date_post'] = date_list
df_ads['car_id'] = car_id_list
df_ads['user_id'] = user_id_list
df_ads['location_id'] = loc_id_list
```
